```python
import pandas as pd

para_predict = pd.read_excel('para_predict.xlsx')
para_predict
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(figsize=(20,10))
plt.xlabel('Serie temporal',fontsize = 20)
ax.set_ylabel('Unidades',fontsize = 20)
ax.set_title('predicción',fontsize = 40)

times = np.arange(np.datetime64('2022-10-27'), np.datetime64('2022-11-16'))

plt.grid()
plt.plot(times,para_predict["unidad"], linewidth=5.0,color='tab:orange' ,label='PREDICT')
plt.bar(times,para_predict["unidad"], linewidth=0.7 , width=0.1,label='DIAS')
ax.legend(fontsize = 20)
plt.savefig("static\pred.jpg")
```
